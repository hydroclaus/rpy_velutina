```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
fname = r"../data/sensor_data_2026-05-14 10:34:34.csv"
```

```python
data = pd.read_csv(fname)
```

```python
data.index = pd.to_datetime(data['datetime'])
```

```python
# t0 = pd.Timestamp('2026-05-13 20:30:00')
# data.index = t0 + pd.to_timedelta(data['datetime'], unit='s')
# data.index.name = 'timestamp'

variables = ['CO2 [PPM]', 'Temp [C]', 'relHum [%]']
ax = data.plot(y=variables, lw=2, subplots=True, grid=True)
for a, lbl in zip(ax, variables):
    a.set_ylabel(lbl)

now = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
plt.tight_layout()
plt.savefig(f'../figs/sensor_data_plot_{now}.png')
```
